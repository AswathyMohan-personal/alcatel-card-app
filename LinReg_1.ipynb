```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression

housing = fetch_california_housing(as_frame = True)
df = housing.frame #contains all columns

#Q.1
print(f"Feature Name: {housing.feature_names}")

#Q.2
input_feature =  df[["MedInc"]] #double bracket is used to get result as dataframe instead of data series
target= df[["MedHouseVal"]]

print(f"Input feature's shape {input_feature.shape}")
print(f"Output label's shape {target.shape}")

#Q.3
plt.figure(figsize=(10,6))
plt.scatter(input_feature,target,alpha=0.6)
plt.xlabel("Median Income - MedInc")
plt.ylabel("Median House Value - MedHouseVal")
plt.title("Scatter Plot")
plt.show()

#Q.4
lin_reg = LinearRegression()
lin_reg.fit(input_feature, target)

print(f"Intercept {lin_reg.intercept_}")
print(f'Coefficent {lin_reg.coef_}')


plt.figure(figsize=(10,6))
plt.scatter(input_feature,target,alpha=0.6)
plt.plot(input_feature,lin_reg.predict(input_feature),color='red')
plt.xlabel("Median Income - MedInc")
plt.ylabel("Median House Value - MedHouseVal")
plt.title("Scatter Plot with regression line")
plt.show()
```
