```python
import pandas as pd

df = pd.read_csv("/content/dataset.csv",index_col=0) #using first col as index
print(df.shape) #number of rows and cols
df.head(10)
```

```python
print(df.columns)
df.info()
```

```python
#check of duplicates
duplicate_count = df.duplicated().sum()
print(duplicate_count)
```

```python
#find missing values
missing_values = df.isna().sum()
#print(missing_values.sum())
if missing_values.sum() == 0:
  print("no missing values")
else:
  for col,count in missing_values.items():
    if count > 0:
      percentage_missing = (count/len(df))*100
      print(f'{col}: {percentage_missing:.lf}%')
```

```python
#check for class imbalance
class_count = df['diagnosis'].value_counts()
print(class_count)


percentage = df["diagnosis"].value_counts(normalize=True)*100
print(f'Count, Percentage : {class_count[0]}, {percentage[0]:.1f}%')
print(f'Count, Percentage : {class_count[1]}, {percentage[1]:.1f}%')

#this stage helps to determine what kind of evaluation matrix we'll be focussing on
#if there is huge class imbalance we;ll use precision or recall
```
